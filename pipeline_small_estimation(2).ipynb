```python
import pandas as pd
```

```python
# fetch data from the file
def read_data(path):
    all = []
    unique = []
    with open(path) as f:
        # read the file line by line
        lines = f.readlines()
        # remove the whitespace characters like `\n` at the end of each line
        lines = list(map(lambda s: s.strip(), lines))
        # split the list into chunks of 4 lines
        aa = [lines[i:i + 4] for i in range(0, len(lines), 4)]
        # getting all unique group names
        for i in aa[:-1]:
            if i[2] not in unique:
                unique.append(i[2])
        # assigning each group to a list
        for j in unique:
            depe = [item for item in aa[:-1] if item[2] == j]
            all.append(depe)
        return all
        
# explanation of the read_data function
# the function takes the path of the file as an argument
# it reads the file line by line and removes the whitespace characters
# it splits the list into chunks of 4 lines
# it gets all unique group names
# it assigns each group to a list
# it returns the list of lists
```

```python
def get_dependent(list2):
    # getting the dependent tasks
    dependents=[]
    for d in list2:
        dependents.append(d[0])
    return dependents

# explanation of the function
# the function takes the list of lists as an argument
# it gets the dependent tasks
# it returns the list of dependent tasks
```

```python
def get_lists(i,list1,list2):
    independent_elements = []
    all_elements=[]
    dependent_elements=[]
    # check if the element is dependent or independent
    for m in range(len(i)):
        l1 = []
        # if feature is independent then add it to the independent list
        if i[m][3] == '':
            independent_elements.append(i[m][0])
            # list 1 contains the independent elements and their repetitions based on execution time
            for n in range(int(i[m][1])):
                l1.append(i[m][0])
        else:
            # if feature is dependent then add it to the dependent list
            dependent_elements.append(i[m][0])
            list2.append(i[m])
        # if the list is not empty then add it to the list1
        if len(l1) > 0:
            list1.append(l1)
        # add all the elements to the all_elements list
        all_elements.append(i[m][0])
    return list1,list2,independent_elements,dependent_elements,all_elements

# explanation of the function
# the function takes the list of lists, list1 and list2 as arguments
# it checks if the element is dependent or independent
# if feature is independent then add it to the independent list
# if feature is dependent then add it to the dependent list
# if the list is not empty then add it to the list1
# it returns the list1,list2,independent_elements,dependent_elements,all_elements
```

```python
def check_dependencies(i,independent_list):
    depend=i[3].split(',')
    # check if the dependent elements are present in the independent list
    total_length=len(depend)
    status=[]
    max_length=[]
    for d in depend:
        for m in independent_list:
            if d in m:
                # if the dependent element is present in the independent list then append true to the status list
                status.append('true')
                max_length.append(len(m))
            else:
                # if the dependent element is not present in the independent list then append false to the status list
                status.append('false')
    # if all the dependent elements are present in the independent list then return true
    value=''
    occurrences = status.count("true")
    if occurrences==total_length :
        ind=max_length.index(max(max_length))
        value=depend[ind]
        status=True

    else:
        status=False
    return status,value

# explanation in detain check_dependencies function
# the function takes the list of lists and independent list as arguments
# it checks if the dependent elements are present in the independent list
# if the dependent element is present in the independent list then append true to the status list
# if the dependent element is not present in the independent list then append false to the status list
# if all the dependent elements are present in the independent list then return true
# it returns the status and value

```

```python
def add_value(data,value,independent_length):
    for n in independent_length:
        if value in n:
            for x in range(int(data[1])):
                n.append(data[0])
    return independent_length

# explanation of the function
# the function takes the data,value and independent_length as arguments
# it adds the value to the independent list
# it returns the independent_length

```

```python
def checkpre(ind,total_list,independent_list):
    new=[]
    for index in ind:
        if index[3]!='':
            depend_var = index[3].split(',')
            # update string to list
            index[3]=index[3].split(',')
            for m in depend_var:
                # if the dependent variable is not in the independent list then remove it from the dependent list
                if m not in total_list:
                    index[3].remove(m)
        new.append(index)
    # if the dependent list is empty then add it to the independent list
    for c in new:
        if type(c[3]) == list and c[3]!=[]:
            c[3]=c[3][0]
        if c[3]==[]:
            lis=[]
            for x in range(int(c[1])):
                lis.append(c[0])
            independent_list.append(lis)
            ind.remove(c)
    depr=[]
    for b in ind:

        depr.append(b[0])
    return independent_list,ind,list(set(depr))

# explanation in detail checkpre function
# the function takes the ind,total_list and independent_list as arguments
# it checks if the dependent variable is not in the independent list then remove it from the dependent list
# if the dependent list is empty then add it to the independent list
# it returns the independent_list,ind,depr
```

```python
# get the final result
def get_data(dependent_list,independent_list,dependent_elements,total_list):
    if dependent_elements==[]:
        return independent_list
    result1=''
    # check if the dependent elements are present in the independent list
    while dependent_elements!=[]:
         for i in dependent_list:
            status, value = check_dependencies(i, independent_list)
            if status == True and value != '':
                result1 = add_value(i, value, independent_list)
                dependent_elements.remove(i[0])
                if dependent_elements==[]:
                    break
    return result1

# explanation in detail how to get the final result
# the function takes the dependent_list,independent_list,dependent_elements,total_list as arguments
# it checks if the dependent elements are present in the independent list
# if the dependent elements are present in the independent list then add it to the independent list
# it returns the result
```

```python
# there for bug in this function in for loop. previously it was for i in all[:-1] so it was not taking the last group in the list. so now i changed it to for i in all: .
def main(all,output_file):
    data=[]
    # iterate over each group
    for i in all:
        feature=i[0][2]
        list1 = []
        list2 = []
        # get the dependent elements, independent elements and all elements in a list
        independent_list, dependent_list, independent_elements,dependent_elements,all_elements = get_lists(i,list1,list2)
        if all.index(i)!=0:
            independent_list,dependent_list,dependent_elements=checkpre(dependent_list,all_elements,independent_list)
        # get the final result
        result=get_data(dependent_list,independent_list,dependent_elements,all_elements)
        # get the dependent tasks
        maxLength = max(len(x) for x in result)
        # append None to the list if the length is less than max length
        for item in result:  # for each item in the list
            while len(item) < maxLength:  # while the item length is smaller than 3
                item.append(None)
        function=[]
        for f in range(maxLength):
            function.append(feature)
        result.append(function)
        df=pd.DataFrame(result)
        previous_df = df.T
        data.append(previous_df)
    dff = pd.concat(data,ignore_index=True)
    dff.to_csv(str(output_file))
    return dff

# explanation mains function and how it works
# the function takes the all and output_file as arguments
# it iterates over each group
# it gets the dependent elements, independent elements and all elements in a list
# it checks if the dependent elements are present in the independent list
# it gets the final result
# it gets the dependent tasks
# it appends None to the list if the length is less than max length
# it returns the dff

```
